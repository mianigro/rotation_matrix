# Rotation of a point around the origin (0,0)

***

## Rotation matix in 2D

$ \mathbf{Point} = \left( \begin{array}{cc} x  \\ y \end{array}\right)$


$ \mathbf{Rotation(\theta)} = \left( \begin{array}{cc} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right)$


$ \mathbf{Point\prime} = \left( \begin{array}{cc} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right) . \left( \begin{array}{cc} x  \\ y \end{array}\right)$

## Rotation equation for points x and y

$x\prime = x \cos\theta - y \sin\theta$

$y\prime = x \sin\theta + y \cos\theta$


***

## Algorithm

### 1. Determine the initial point to rotate, given as (1,0) and set the origin as (0,0)
### 2. Apply the above rotation formula to each value of $\theta$ to the $x$ and the $y$ to determine $x\prime$ and $y\prime$.
### 3. Display the original point and the new point as vectors with the tail at the origin (0,0)

***

## Importing modules

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Rotation function

This function takes the input $\theta$ in radians and completes the rotation function based on the supplied formulae, rotating the point around the origin (0, 0).

The initial point (1,0) is set as (x, y) and applied to the formulae, which creates the rotated point x$\prime$ and y$\prime$, which is rotated around the origin. The rotated point, (x$\prime$, y$\prime$) is displayed in a graph with the original point (x, y). All of the points use the origin (0,0) as the tail of the vector as that is the point at which they rotate around.

#### Rotation Formula

$x\prime = x \cos\theta - y \sin\theta$

$y\prime = x \sin\theta + y \cos\theta$

```python
def point_rotation(theta):
    # Set origin at (0,0) and point to rotate at (1,0)
    p_o = (0,0)
    p = (1, 0)
    
    # Use the x and y rotation formulas to rotate around the origin    
    p_prime = (round((p[0]*math.cos(theta)-p[1]*math.sin(theta)), 2), round((p[0]*math.sin(theta)+ p[1]*math.cos(theta)), 2))
    
    # Printing outputs to the user to view the relevant information
    print("------------------------------------------")
    print(f"Point = {p}")
    print(f"Rotation anticlockwise in degrees = {round((theta * 180 / math.pi), 2)}")
    print(f"Point' = {p_prime}")
    
    # Display the rotation as vectors from the origin and setup cartesian coordinates
    plt.rcParams["figure.figsize"] = (5,5)
    ax = plt.axes()
    ax.arrow(p_o[0],p_o[1], p[0],p[1] , head_width=0.1, head_length=0.1)
    ax.arrow(p_o[0],p_o[1], p_prime[0],p_prime[1] , head_width=0.1, head_length=0.1)
    ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    x_ticks = np.arange(-2, 2, 0.25)
    y_ticks = np.arange(-2, 2, 0.25)
    ax.set_xticks(x_ticks[x_ticks != 0])
    ax.set_yticks(y_ticks[y_ticks != 0])
    ax.set_xticks(np.arange(-2, 2, 0.25), minor=True)
    ax.set_yticks(np.arange(-2, 2, 0.25), minor=True)

    plt.ylim(-1.2,1.2)
    plt.xlim(-1.2,1.2)
    
    plt.show()

    
```

## Rotation for values of $\theta$

This function loops through a series of values for $\theta$ in radians to illustrate the different angles.

```python
thetas = [math.pi/8, math.pi/4, math.pi/2, 3*math.pi/4]

for theta_in in thetas:
    point_rotation(theta_in)
```
